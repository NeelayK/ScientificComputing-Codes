```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def svd_image_compression(image_array, k):
    U, S, Vt = np.linalg.svd(image_array, full_matrices=False)
    U_k = U[:, :k]
    S_k = np.diag(S[:k])
    Vt_k = Vt[:k, :]
    reconstructed_image = U_k @ S_k @ Vt_k
    return reconstructed_image

image_path = "input.jpeg"
image = Image.open(image_path).convert('L')
og = np.array(image, dtype=float)
#k = int(input("Enter k:"))
k5 = svd_image_compression(og, 5)
k10 = svd_image_compression(og, 10)
k20 = svd_image_compression(og, 20)
k40 = svd_image_compression(og, 40)
k60 = svd_image_compression(og, 60)

plt.figure(figsize=(40, 20))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(og, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title(f"Compressed Image (k=5)")
plt.imshow(k5, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title(f"Compressed Image (k=10)")
plt.imshow(k10, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.title(f"Compressed Image (k=20)")
plt.imshow(k20, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title(f"Compressed Image (k=40)")
plt.imshow(k40, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title(f"Compressed Image (k=60)")
plt.imshow(k60, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```
